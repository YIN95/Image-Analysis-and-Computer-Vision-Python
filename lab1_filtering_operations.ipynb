# Properties of discrete Fourier transform

## The continuous and discrete Fourier transform

From the definitions used during the leactures the continuous Fourier transform $\hat{f}:\mathbb{R}^2 \rightarrow  \mathbb{C}$ of a two-dimensional signal $\hat{f}:\mathbb{R}^2 \rightarrow  \mathbb{R}$ is



> $\hat{f}(\omega)=\mathcal{F}_{C}(f)(\omega)=\int_{x\in \mathbb{R}^2}f(x)e^{(-i\omega^Tx)}dx$







```
import numpy as np
import matplotlib.pyplot as plt

def showgray(F:np.array) -> np.array:
    F = F - np.min(F)
    Fmax = np.max(F)
    F /= Fmax
    
    return F
    
def fftwave(u:int, v:int, sz:int=128):
    Fhat = np.zeros((sz, sz))
    Fhat[u, v] = 1
    Fhat_fftshift = np.fft.fftshift(Fhat)
    F = np.fft.ifft2(Fhat)
    Freal = np.real(F)
    Fimag = np.imag(F)
    Fabs = np.abs(F)
    Fangle = np.angle(F)
    
    '''
    Question: 
        What is done by these instructions?
            uc = u if u <= sz/2 else uc = u -sz
            vc = v if v <= sz/2 else vc = v -sz
    Answer:
        Write your answer here ...
    '''
    
    uc = u if u <= sz/2 else u - sz
    vc = v if v <= sz/2 else v - sz
    
    wavelength = 0.0 # Replace by correct expression
    amplitude = 0.0 # Replace by correct expression
    
    # Visualization
    fig, axes = plt.subplots(1, 2)
    
    axes[0].imshow(showgray(Fhat), cmap='gray')
    axes[0].title.set_text('Fhat:(u, v) = ({}, {})'.format(u, v))
    axes[0].axis('off')
    
    axes[1].imshow(showgray(Fhat_fftshift), cmap='gray')
    axes[1].title.set_text('(uc, vc) = ({}, {})'.format(uc, vc))
    axes[1].axis('off')
    
    fig, axes = plt.subplots(1, 2)
    axes[0].imshow(showgray(Freal), cmap='gray')
    axes[0].title.set_text('real(F)')
    axes[0].axis('off')
    
    axes[1].imshow(showgray(Fimag), cmap='gray')
    axes[1].title.set_text('imag(F)')
    axes[1].axis('off')
    
    fig, axes = plt.subplots(1, 2)
    axes[0].imshow(showgray(Fabs), cmap='gray')
    axes[0].title.set_text('amplitude = {}'.format(amplitude))
    axes[0].axis('off')

    axes[1].imshow(showgray(Fangle), cmap='gray')
    axes[1].title.set_text('wavelength = {}'.format(wavelength))
    axes[1].axis('off')
```
